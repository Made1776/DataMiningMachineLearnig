## (b) Create a 2-dimensional array of 3 rows and 4 columns with random integer values. Then, calculate the sum of each row and column.

```python
# Crear la matriz aleatoria
matriz = np.random.randint(1, 10, size=(3, 4))
print("Matriz:")
print(matriz)

# Calcular la suma de cada fila
suma_filas = np.sum(matriz, axis=1)
print("Suma de cada fila:", suma_filas)
#print()

# Calcular la suma de cada columna
suma_columnas = np.sum(matriz, axis=0)
print("Suma de cada columna:", suma_columnas)
```
